```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'date': ['2024-03-01','2024-03-01','2024-03-01','2024-03-01','2024-03-01'],
    'datetime': ['2024-03-01 10:15:50.520','2024-03-01 12:19:22.539',
                 '2024-03-01 12:20:18.089','2024-03-01 13:46:33.006',
                 '2024-03-01 13:48:14.626'],
    'cash_type': ['card','card','card','card','card'],
    'card': ['ANON-0000-0000-0001','ANON-0000-0000-0002',
             'ANON-0000-0000-0002','ANON-0000-0000-0003',
             'ANON-0000-0000-0004'],
    'money': [38.7, 38.7, 38.7, 28.9, 38.7],
    'coffee_name': ['Latte','Hot Chocolate','Hot Chocolate','Americano','Latte']
}
```

```python
v = pd.DataFrame(data)
```

```python
v['datetime'] = pd.to_datetime(v['datetime'])
v['date'] = pd.to_datetime(v['date'])
```

```python
cols_to_encode = ['cash_type', 'card', 'coffee_name']
```

```python
encoder = LabelEncoder()
for col in cols_to_encode:
    v[col] = encoder.fit_transform(v[col])

print("\nEncoded DataFrame:\n", v)
print("\nData types after encoding:\n", v.dtypes)
```

```python
v.head()
```

```python
v['date'] = pd.to_datetime(v['date'])
```

```python
daily_sales = v.groupby('date')['money'].sum().reset_index()
```

```python
v['hour'] = v['datetime'].dt.hour
hourly_sales = v.groupby('hour')['money'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=hourly_sales, x='hour', y='money', palette='viridis')
plt.title('Sales by Hour of Day')
plt.show()
```

Sales are increasing at 12 o'clock.decreasing at 10 o'clock

```python
v['weekday'] = v['date'].dt.day_name()
weekday_sales = v.groupby('weekday')['money'].sum().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=weekday_sales, x='weekday', y='money', palette='coolwarm')
plt.title('Sales by Day of the Week')
plt.xticks(rotation=45)
plt.show()
```

sales are increasing on Friday.

```python
sales_pivot = v.groupby(['hour','weekday'])['money'].sum().reset_index()
sales_pivot = sales_pivot.pivot(index='hour', columns='weekday', values='money')
sales_pivot = sales_pivot.reindex(columns=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']).fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(sales_pivot, cmap='YlGnBu')
plt.title('Heatmap of Sales by Hour and Weekday')
plt.show()
```

The heat map increases at 12 o'clock. Decreses in 10 o'clock
